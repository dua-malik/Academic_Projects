```python
# Abstract representation of players and empty space -- an empty board will be a list of lists of 0s (empty)
player_1 = 1
player_2 = 2
empty = 0

# When we start drawing the board, we will place "pieces" down for each player. Player 1 will put down "X", 
# Player 2 will put down "O", and an empty space will be drawn as a space character
player_1_piece = "X"
player_2_piece = "O"
empty_space = " "

# This dictionary will be used to draw the game board. 
space_character = {player_1: player_1_piece,
                  player_2: player_2_piece,
                  empty: empty_space}
```

```python
def draw_board(board):
    num_rows = len(board)
    num_cols = len(board[0])
    
    row_names = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    row_map = dict(zip(row_names,range(num_rows)))
    col_names = list(map(str,range(1,num_cols+1)))
    col_map = dict(zip(col_names,range(num_cols)))
    
    print("   ", row_names[0], end = "")
    for i in range(num_cols-1):
        print("  ", row_names[i+1], end = "")
    print("\n   ---", end = "")
    print(" ---"*(num_cols-1))
    
    for i in range(num_rows):
        print(col_names[i],"|",end=" ")
        for j in range(num_cols):
            print(space_character[board[i][j]],end=" ")
            print("|",end=" ")
            
        print("\n   ---", end ="")  
        print(" ---"*(num_cols-1))
def make_board(n):
    if isinstance(n,int) == False:
        print("Error: Board size must be an integer.")
    else: 
        return [[empty]*n for i in range(n)]

def show_board(board):
    for i in board:
        print(i)
        
        
def make_move(board,player,location):
    row_names = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    col_names = list(range(len(row_names)))
    move_dictionary = dict(zip(row_names, col_names))
    x = list(location)
    x[1] = int(x[1]) - 1
    x[0] = move_dictionary[x[0]]
    
    if x[0] not in range(len(board)) or x[1] not in range(len(board)):
        print("Location is invalid.")
        return False 
    
    if board[x[1]][x[0]] == empty:
        board[x[1]][x[0]] = player
        return True 

def valid_move(board,player,valid):
    while not valid:
        valid_location = input("input valid location.")
        valid = make_move(board,player,valid_location) 

def game_won(board):
    x, y = diagonal_equal(board)
    if x == False:
        x, y = diagonal2_equal(board)
    if x == False:
        x, y = row_equal(board)
    if x == False:
        x, y = col_equal(board)
    if x == True:
        return y
    for i in range(len(board)):
        for j in range(len(board)):
            if board[i][j] == empty:
                return -1
    return 0
```

```python
def ttt_game():
    print("Welcome to TicTacToe!")
    print("-----------------------------")
    
    size_0 = int(input("input board size:"))
    board = make_board(size_0)
    draw_board(board)
    player = player_1
    
    state = -1
    
    # how to run game:
    while state == -1:
        
        # want to draw the board so the player can see what they are doing
        print('Player', player, 'move:', end = " ")
        player_move = input("")
        is_valid = make_move(board,player,player_move)
        valid_move(board, player, is_valid)

        state = game_won(board)
        if state == 1:
            print("Player 1 has won")
        elif state == 2:
            print("Player 2 has won")
        elif state == 0:
            print("Game is a draw")
        # assume that the game returns False -- now you need to switch players 
        if player == player_1:
            player = player_2
        else:
            player = player_1

        draw_board(board)
    
# checking if diagonal, rows, and columns are equal 
def diagonal_equal(board):
    diag = list()
    for i in range(len(board)):
        if board[i][i] != 0:
            diag.append(board[i][i])
    if diag == [board[0][0]]*len(board):
        return True, diag[0]
    else:
        return False, 0
    
def diagonal2_equal(board):
    diag2= list()
    for i in range(len(board)):
        if board[i][i] != 0:
            diag2.append(board[i][-1-i])
    if diag2 == [board[0][-1]]*len(board):
        return True, diag2[0]
    else:
        return False, 0 

            
def row_equal(board):
    for i in range(len(board)):
        row = list()
        for j in range(len(board)):
            if board[i][i] != 0:
                row.append(board[i][j])
        if row == [board[i][j]]*len(board):
            return True, row[0]
    return False, 0 

def col_equal(board):
    for i in range(len(board)):
        col = list()
        for j in range(len(board)):
            if board[i][i] != 0:
                col.append(board[j][i])
        if col == [board[j][i]]*len(board):
            return True, col[0]
    return False, 0
```
