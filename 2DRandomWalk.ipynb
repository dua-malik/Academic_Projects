## Importing Packages:

```python
import numpy as np
import matplotlib as plt
from matplotlib import pyplot as plt
```

## Random Walk Simulator in 1D:

```python
def randomwalk(num_steps,start_position):
    final_position = start_position + np.sum((np.random.rand(num_steps)))
    return final_position

def simulaterandomwalk(num_sims,num_steps,start_position):
    final_step = np.zeros(num_sims)
    for i in range(num_sims):
        final_step[i] = randomwalk(num_steps,start_position)
    return final_step


x = simulaterandomwalk(15,10,0)
print(x)
```

```python
randomwalkhist = simulaterandomwalk(10000,10,0)
plt.hist(randomwalkhist)

plt.show()
```
