# Principal Component Analysis on Wine Data

```python
# importing packages 
import numpy as np 
import matplotlib
from matplotlib import pyplot as plt
import sklearn as skl 
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
```

#### Importing  wine dataset, which is a classification dataset with 13 dimensions (i.e., the dataset is 13 dimensional). It has 3 classes. 

```python
winedata = skl.datasets.load_wine()
```

#### X stands for our feature matrix. This is the matrix with columns as features and rows as samples. y stands for the labels (we have 3 different labels/classes). Assigning the winedata ['data'] and winedata ['target'] to X and y.

```python
X = winedata['data']
y = winedata['target']
print(X,y)
```

####  Using the shape method to check the shape of the feature matrix. 

```python
print(X.shape)

# there are 178 samples in our dataset and 13 features
```
